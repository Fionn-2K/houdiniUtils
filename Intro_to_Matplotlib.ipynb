```python
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as np
```

```python
X = np.arange(0,100,5)
Y = np.arange(100,200,5)
```

```python
plt.scatter(X,Y,c=Y, cmap="RdYlBu")
```

```python
n_samples = 1000

X, y = datasets.make_circles(n_samples, noise=0.03, random_state=12)
```

```python
plt.scatter(x=X[:,0], y=X[:,1], c=y, cmap=plt.cm.RdYlBu);
```

```python
NUM_CLASSES = 4
NUM_FEATURES = 2

X_blob, y_blob = datasets.make_blobs(
                                    n_samples=2000, 
                                    n_features=NUM_FEATURES, 
                                    centers=NUM_CLASSES, 
                                    cluster_std=1.3,
                                    random_state=12
                                    )
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X_blob[:,0], X_blob[:,1], c=y_blob, cmap=plt.cm.RdYlBu)
```

```python
x = np.linspace(0,20,5)
y = x ** 2
```

```python
plt.plot(x,y,"r")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Test data")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.subplot(1,2,2)
plt.plot(y,x,"g*-")
```
