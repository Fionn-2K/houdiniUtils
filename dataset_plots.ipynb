```python
import pandas as pd
import numpy as np
```

```python
# First dataset. Best Video Games of All Time (https://www.kaggle.com/datasets/faisaljanjua0555/best-video-games-of-all-time)
vg_data = pd.read_csv("vg_data.csv")
vg_data
```

```python
# Tidy-up data. Remove columns with missing values
vg_data = vg_data.dropna(axis=1)
vg_data = vg_data.rename(columns={'Title' : 'Description'})
vg_data.head()
```

```python
vg_data.info()
```

```python
vg_sales = pd.read_csv("vg_sales.csv")
vg_sales.head()
```

```python
# Second dataset. Global Video Games Sales (https://www.kaggle.com/datasets/thedevastator/global-video-game-sales)
#vg_sales["Platform"].replace("NES", "Nintendo", inplace=True)
vg_sales["Platform"] = vg_sales["Platform"].replace("NES", "Nintendo")
vg_sales["Platform"] = vg_sales["Platform"].replace("N64", "Nintendo 64")
vg_sales["Platform"] = vg_sales["Platform"].replace("GC", "GameCube")
vg_sales["Platform"] = vg_sales["Platform"].replace("WiiU", "Wii U")
vg_sales["Platform"] = vg_sales["Platform"].replace("GB", "Game Boy")
vg_sales["Platform"] = vg_sales["Platform"].replace("GBA", "Game Boy Advance")
vg_sales["Platform"] = vg_sales["Platform"].replace("PS", "PlayStation")
vg_sales["Platform"] = vg_sales["Platform"].replace("PS2", "PlayStation 2")
vg_sales["Platform"] = vg_sales["Platform"].replace("PS2", "PlayStation 3")
vg_sales["Platform"] = vg_sales["Platform"].replace("PS2", "PlayStation 4")
vg_sales["Platform"] = vg_sales["Platform"].replace("PS2", "PlayStation 5")
vg_sales["Platform"] = vg_sales["Platform"].replace("XB", "Xbox")
vg_sales["Platform"] = vg_sales["Platform"].replace("X360", "Xbox 360")
vg_sales["Platform"] = vg_sales["Platform"].replace("XOne", "Xbox One")
vg_sales["Platform"] = vg_sales["Platform"].replace("DC", "Dreamcast")
vg_sales
```
