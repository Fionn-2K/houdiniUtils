```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
x = np.linspace(-10,10,1000)

y = x**2 + 2*x +2

fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
fig ,ax = plt.subplots(figsize=(6,4))

ax.plot(x, y+1, color="red", linewidth=0.2)
ax.plot(x, y+2, color="red", linewidth=0.5)
ax.plot(x, y+3, color="red", linewidth=0.1)

ax.plot(x, y+5, color="green", linestyle="-")
# ax.plot(x, y+6, color="green", linestyle="_,")
# ax.plot(x, y+7, color="green", linestyle=":")

line, = ax.plot(x, y+8, color="black", lw=1.5)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x, y**2, x, np.exp(x))
ax.set_title("Numbers")
ax.set_yticks([0,50,100,150])

from matplotlib import ticker

ft = ticker.ScalarFormatter(useMathText=True)
ft.set_scientific(True)
ft.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(ft)
ax.yaxis.labelpad = 15
ax.xaxis.labelpad = 10
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,3))

axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color="b", alpha=0.5, linestyle="--", lw=0.5)
```

```python
n = np.arange(1,50)
x = np.linspace(-0.5, 1., 50)
```

```python
fig, axes = plt.subplots(1,4, figsize=(12,3))

axes[0].scatter(x, x + 0.25 * np.random.randn(len(x)))
axes[0].set_title("Scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("Step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("Bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5)
axes[3].set_title("Fill");
```

```python
X,Y = np.meshgrid(np.linspace(0,2*np.pi,300),np.linspace(0,2*np.pi,300))
Z = 2 * np.cos(X)
```

```python
fig = plt.figure(figsize=(10,5))

ax = fig.add_subplot(1,2,1, projection="3d")
p = ax.plot_surface(X,Y,Z, rstride=4, cstride=4, linewidth=0)

ax = fig.add_subplot(1,2,2, projection="3d")
p = ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap=matplotlib.cm.viridis, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5);

cset = ax.contour(X,Y,Z, zdir="z", offset=-np.pi, cmap="coolwarm")
cset = ax.contour(X,Y,Z, zdir="x", offset=-np.pi, cmap="coolwarm")
cset = ax.contour(X,Y,Z, zdir="y", offset=-np.pi, cmap="coolwarm")
```
